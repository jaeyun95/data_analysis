```python
import matplotlib.pyplot as plt
import numpy as np

## test
plt.plot([1,2,3,4,5,4,3,2,1])
plt.show()

```

```python
## test
x = np.arange(0,8,0.01)
y = np.sin(x)
plt.plot(x,y)
plt.show()

```

```python
## test
x = np.arange(0,8,0.01)
y = np.sin(x)
plt.plot(x,y)
plt.grid()
plt.xlabel("time")
plt.ylabel("Amplitude")
plt.title('Example of sinewave')
plt.show()

```

```python
## test
x = np.arange(0,8,0.01)
y = np.sin(x)
y_ = np.cos(x)
plt.plot(x,y)
plt.plot(x,y_)
plt.grid()
plt.xlabel("time")
plt.ylabel("Amplitude")
plt.title('Example of sinewave')
plt.show()

```

```python
## test
x = np.arange(0,8,0.01)
y = np.sin(x)
y_ = np.cos(x)
plt.plot(x,y,label='sin')
plt.plot(x,y_,label='cos')
plt.grid()
plt.legend()
plt.xlabel("time")
plt.ylabel("Amplitude")
plt.title('Example of sinewave')
plt.show()

```

```python
## test
x = np.arange(0,8,0.01)
y = np.sin(x)
y_ = np.cos(x)
plt.plot(x,y,label='sin', color="green") #color change
plt.plot(x,y_,label='cos', lw=3) #bold
plt.grid()
plt.legend()
plt.xlabel("time")
plt.ylabel("Amplitude")
plt.title('Example of sinewave')
plt.show()

```

```python
## test
x = np.arange(0,8,0.01)
y = np.sin(x)
y_ = np.cos(x)
plt.plot(x,y,label='sin', linestyle='dashed') 
plt.plot(x,y_,label='cos')
plt.grid()
plt.legend()
plt.xlabel("time")
plt.ylabel("Amplitude")
plt.title('Example of sinewave')
plt.show()
```
