```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

x = np.linspace(0,14,100)
y1 = np.sin(x)
y2 = 2*np.sin(x+0.5)
y3 = 3*np.sin(x+1.0)
y4 = 4*np.sin(x+1.5)

plt.plot(x,y1,x,y2,x,y3,x,y4)
plt.show()

sns.set_style("whitegrid")
plt.plot(x,y1,x,y2,x,y3,x,y4)
plt.show()
```

```python
import pandas as pd
import numpy as np

df = pd.read_csv("/content/drive/MyDrive/data/DataScience-master/DataScience-master/data/02. Final_crime_in_Seoul.csv")
df = pd.pivot_table(df,index="구별")
df.head()
```

```python
sns.pairplot(df,vars=["강도","살인","폭력"], kind="reg", size=3)
plt.show()
```

```python
sns.pairplot(df, x_vars=["인구수","CCTV"],y_vars=["살인","강도"],kind="reg",size=3)
plt.show()
```

```python
sns.pairplot(df, x_vars=["인구수","CCTV"],y_vars=["살인검거율","강도검거율"],kind="reg",size=3)
plt.show()
```

```python
max_ = df["검거"].max()
df["검거"] = df["검거"] / max_*100
sort_df = df.sort_values(by="검거",ascending=False)
sort_df.head()
```

```python
col = ["강간검거율","강도검거율","살인검거율","절도검거율","폭력검거율"]

plt.figure(figsize=(10,10))
sns.heatmap(sort_df[col], annot=True, fmt="f", linewidths=.5, cmap="Blues")
plt.show()
```

```python
col = ["강간","강도","살인","절도","폭력","범죄"]

df["범죄"] = df["범죄"] / 5
sort_df = df.sort_values(by="범죄",ascending=False)
plt.figure(figsize=(10,10))
sns.heatmap(sort_df[col], annot=True, fmt="f",linewidths=.5, cmap="YlGnBu")
plt.show()
```
