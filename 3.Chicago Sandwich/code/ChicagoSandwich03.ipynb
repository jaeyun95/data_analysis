```python
import pandas as pd
from bs4 import BeautifulSoup
from urllib.request import urlopen,Request

data_frame = pd.read_csv("/content/drive/MyDrive/data/best_sandwiches_chicago.csv",index_col=0)
data_frame.head()
```

```python
request = Request(data_frame["URL"][0], headers={'User-Agent': 'Mozilla/5.0'})
print(data_frame["URL"][0])
html = urlopen(request).read()
soup_tmp = BeautifulSoup(html,"html.parser")
soup_tmp
```

```python
soup_tmp.find('p','addy').get_text().split()
```

```python
price = []
address = []

for n in data_frame.index[:3]:
  request = Request(data_frame["URL"][n], headers={'User-Agent': 'Mozilla/5.0'})
  html = urlopen(request).read()
  soup_tmp = BeautifulSoup(html,"html.parser")
  gettings = soup_tmp.find('p','addy').get_text()
  price.append(gettings.split()[0][:-1])
  address.append(' '.join(gettings.split()[1:-2]))
    
```

```python
import tqdm
```

```python
from tqdm import tqdm_notebook

price = []
address = []

for n in tqdm_notebook(data_frame.index):
  request = Request(data_frame["URL"][n], headers={'User-Agent': 'Mozilla/5.0'})
  html = urlopen(request).read()
  soup_tmp = BeautifulSoup(html,"html.parser")
  gettings = soup_tmp.find('p','addy').get_text()
  price.append(gettings.split()[0][:-1])
  address.append(' '.join(gettings.split()[1:-2]))
```

```python
data_frame["Price"] = price
data_frame["Address"] = address

data_frame = data_frame.loc[:,["Rank","Cafe","Menu","Price","Address"]]
data_frame.set_index("Rank",inplace=True)
data_frame.head()
```

```python
data_frame.to_csv("/content/drive/MyDrive/data/best_sandwiches_chicago2.csv",sep=",",encoding="UTF-8")
```
