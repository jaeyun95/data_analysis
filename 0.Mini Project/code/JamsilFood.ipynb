```python
restaurant_score = []
restaurant_name = []
restaurant_point = [] 
restaurant_address = []
restuarant_URL = []

url_base = "https://www.mangoplate.com"
for i in range(10):
  restaurant_name.append(soup.find_all('span','title')[i].h3.string)
  restaurant_score.append(soup.find_all('span','title')[i].a.get_text().split(" ")[-2][:-1])
  restaurant_point.append(soup.find_all('strong','point')[i].span.string)
  restaurant_address.append(soup.find_all('p','etc')[i].string)
  restuarant_URL.append(url_base + soup.find_all('figure','restaurant-item')[i].a["href"])
  
```

```python
data_frame = pd.DataFrame({"score":restaurant_score,"name":restaurant_name,"point":restaurant_point,"address":restaurant_address,"URL":restuarant_URL})
data_frame.set_index("score",inplace=True)
data_frame.head()
```

```python
from tqdm import tqdm_notebook

phone_number = []
opening_hours = []
for url in tqdm_notebook(data_frame["URL"]):
  request = Request(url, headers={'User-Agent': 'Mozilla/5.0'})
  html = urlopen(request).read()
  soup = BeautifulSoup(html,"html.parser")
  phone_number.append(soup.find_all('tr')[1].td.get_text())
  opening_hours.append(soup.find_all('tr')[5].td.get_text())
```

```python
data_frame['phone number'] = phone_number
data_frame['opening hours'] = opening_hours
data_frame.drop(["URL"],axis=1,inplace=True)
data_frame.head()
```

```python
data_frame.to_csv("/content/drive/MyDrive/data/Jamsil_Food.csv",sep=",",encoding="UTF-8")
```

```python
import googlemaps

gmaps_key = "*******************************"
gmaps = googlemaps.Client(key=gmaps_key)

lat = []
lng = []

for address in tqdm_notebook(data_frame["address"]):
    gmaps_output = gmaps.geocode(address)
    location_output = gmaps_output[0].get('geometry')
    lat.append(location_output['location']['lat'])
    lng.append(location_output['location']['lng'])

data_frame["lat"] = lat
data_frame["lng"] = lng

data_frame.head()
```

```python
import folium

mapping = folium.Map(location=[data_frame["lat"].mean(), data_frame["lng"].mean()], zoom_start=15)

for n in data_frame.index:
  folium.Marker([data_frame["lat"][n], data_frame["lng"][n]],popup=data_frame["name"][n]).add_to(mapping)
mapping
```
